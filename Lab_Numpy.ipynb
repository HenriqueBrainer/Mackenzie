# `Numpy`

## Definindo arrays e suas dimensões

````python
import numpy as np

a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim,a.shape)
print(b.ndim,b.shape)
print(c.ndim,c.shape)
print(d.ndim,d.shape)
````

## Indexing

````python
b[0], c[0,1], c[1,2], d[1,1,1]
````

## Slicing `[start:end:step]`

````python
b[2:3], b[2:4], b[2::], b[2:-1], b[::2]
````

## Datatypes

````python
np.array(['love', 'is', 'love']) # Unicode
````

````python
np.array([1, 2, 3], dtype='S')
````

````python
np.array([1, 2, 3], dtype='f')
````

## Iteractions

````python
for elemento in b:
  print(elemento)

print()

for elemento in d:
  print(elemento)
````

## Reshape

````python
d.reshape(-1), d.flatten(), d.ravel()
````

## Operations

````python
a = np.array([1,2,3])
b = np.array([1,-1,1])

a * 3, np.log(a), a + b, b ** 2, a @ b, np.sum(b), np.mean(a), np.std(a), np.sort(b)
````

## Concatenate e Append

````python
a = np.array([1,2,3])
b = np.array([1,-1,1])

c = np.array([[1,2,3], [0, 0, 0]])
d = np.array([[1,-1,1], [1, 2, 3]])

np.concatenate((a,b)), np.concatenate((c,d)), np.concatenate((c,d), axis=1), np.append(a,[4])
````

## Special arrays

````python
np.ones(2), np.zeros([1,2]), np.identity(4), np.random.rand(4), np.random.normal()
````

# Exercício 4

Faça um `cut` da imagem para exibir apenas o rosto da atriz. Exiba depois apenas uma componente das cores e veja o resultado.

````python
M = np.array(img)
M = M[246:397,100:260,:]
Image.fromarray(M)
````

## Alterando o tamanho da imagem

````python
img_ = Image.fromarray(M)
new_size = ( int(img_.size[0] * 2), int(img_.size[1] / 2) )
img_b = img_.resize( new_size, Image.ANTIALIAS)

img_b # .save('output.png')
````
